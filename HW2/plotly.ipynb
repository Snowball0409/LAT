```python
# Question
filter = (data["Category1Title"]=='總計') & (data["Category2Title"]=='學士班')
total_student = data[filter]
display(total_student)

#draw graph
fig = px.line(total_student, x='Period', y='Val', color='Category3Title', title='民國97年至110年大學生各領域人數變化')
fig.show()
```



---


上圖由於資料間數值相差過大，造成數據變化程度不明顯  


> 以下對資料做Preprocessing：
*   使用**Pandas.DataFrame.pct_change()**功能計算變化量 
*   將**變化率放大10倍**使數據變化量明顯




```python
#Preprocess datasets
preprocess = (total_student.drop(labels=["Category1Title","Category2Title","Category4Title","Category5Title"],axis=1).set_index(["Period","Category3Title"]).unstack().pct_change().stack()*10).reset_index()
display(preprocess)
dig = px.line(preprocess, x='Period', y='Val', color='Category3Title',title='民國97年至110年大學生各領域人數變化')
dig.show()
```
